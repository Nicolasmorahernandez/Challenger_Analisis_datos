### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda = tienda["Precio"].sum()
ingreso_total_tienda2 = tienda2["Precio"].sum()
ingreso_total_tienda3 = tienda3["Precio"].sum()
ingreso_total_tienda4 = tienda4["Precio"].sum()
ingresos_totales = [ingreso_total_tienda, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
print("Ingreso total tienda 1: ", ingreso_total_tienda)
print("Ingreso total tienda 2: ", ingreso_total_tienda2)
print("Ingreso total tienda 3: ", ingreso_total_tienda3)
print("Ingreso total tienda 4: ", ingreso_total_tienda4)

tienda['Ingresos Netos'] = tienda['Precio'] - tienda['Costo de envío']
tienda2['Ingresos Netos'] = tienda2['Precio'] - tienda2['Costo de envío']
tienda3['Ingresos Netos'] = tienda3['Precio'] - tienda3['Costo de envío']
tienda4['Ingresos Netos'] = tienda4['Precio'] - tienda4['Costo de envío']
ingresos_netos_tienda = tienda['Ingresos Netos'].sum()
ingresos_netos_tienda2 = tienda2['Ingresos Netos'].sum()
ingresos_netos_tienda3 = tienda3['Ingresos Netos'].sum()
ingresos_netos_tienda4 = tienda4['Ingresos Netos'].sum()
ingresos_netos_totales = [ingresos_netos_tienda, ingresos_netos_tienda2, ingresos_netos_tienda3, ingresos_netos_tienda4]
```

# 4. Productos más y menos vendidos

```python
productos_unicos_tienda = list(set(tienda["Producto"]))
productos_unicos_tienda2 = list(set(tienda2["Producto"]))
productos_unicos_tienda3 = list(set(tienda3["Producto"]))
productos_unicos_tienda4 = list(set(tienda4["Producto"]))
```

```python
lista_ventas_cantitad_tienda = tienda["Producto"].value_counts()
lista_ventas_cantitad_tienda = lista_ventas_cantitad_tienda.sort_values(ascending = False)
lista_ventas_cantitad_tienda2 = tienda2["Producto"].value_counts()
lista_ventas_cantitad_tienda2 = lista_ventas_cantitad_tienda2.sort_values(ascending = False)
lista_ventas_cantitad_tienda3 = tienda3["Producto"].value_counts()
lista_ventas_cantitad_tienda3 = lista_ventas_cantitad_tienda3.sort_values(ascending = False)
lista_ventas_cantitad_tienda4 = tienda4["Producto"].value_counts()
lista_ventas_cantitad_tienda4 = lista_ventas_cantitad_tienda4.sort_values(ascending = False)
```

```python
def mas_vendido(lista):
  productos_mas_vendidos = []
  for i in range(len(lista)):
    conteo = lista.iloc[i]
    max_value = lista.max()
    if conteo == max_value:
        productos_mas_vendidos.append(
            lista.index[i] + " - " + str(conteo)
          )
  return productos_mas_vendidos
```

```python
productos_mas_vendidos_tienda = mas_vendido(lista_ventas_cantitad_tienda)
print(productos_mas_vendidos_tienda)
productos_mas_vendidos_tienda2 = mas_vendido(lista_ventas_cantitad_tienda2)
print(productos_mas_vendidos_tienda2)
productos_mas_vendidos_tienda3 = mas_vendido(lista_ventas_cantitad_tienda3)
print(productos_mas_vendidos_tienda3)
productos_mas_vendidos_tienda4 = mas_vendido(lista_ventas_cantitad_tienda4)
print(productos_mas_vendidos_tienda4)
```

```python
def menos_vendido(lista):
  productos_menos_vendidos = []
  for i in range(len(lista)):
    conteo = lista.iloc[i]
    min_value = lista.min()
    if conteo == min_value:
        productos_menos_vendidos.append(
            lista.index[i] + " - " + str(conteo)
          )
  return productos_menos_vendidos
```

```python
productos_menos_vendidos_tienda = menos_vendido(lista_ventas_cantitad_tienda)
print(productos_menos_vendidos_tienda)
productos_menos_vendidos_tienda2 = menos_vendido(lista_ventas_cantitad_tienda2)
print(productos_menos_vendidos_tienda2)
productos_menos_vendidos_tienda3 = menos_vendido(lista_ventas_cantitad_tienda3)
print(productos_menos_vendidos_tienda3)
productos_menos_vendidos_tienda4 = menos_vendido(lista_ventas_cantitad_tienda4)
print(productos_menos_vendidos_tienda4)
```

```python
def mapa_dispersion(lista):
  plt.figure(figsize=(8,6))
  plt.scatter(lista['lon'], lista['lat'],
              c=lista['Precio'],        # colorea según volumen de ventas
              cmap='viridis',         # paleta de calor
              s=50,                   # tamaño de los puntos
              alpha=0.7)
  plt.colorbar(label='Precio')
  plt.title('Dispersión del precio por Ubicación')
  plt.xlabel('Longitud')
  plt.ylabel('Latitud')
  plt.tight_layout()
  return plt.show()

mapa_dispersion_tienda = mapa_dispersion(tienda)
mapa_dispersion_tienda2 = mapa_dispersion(tienda2)
mapa_dispersion_tienda3 = mapa_dispersion(tienda3)
mapa_dispersion_tienda4 = mapa_dispersion(tienda4)
```

```python
import seaborn as sns
def heatmaps(lista):
  plt.figure(figsize=(8,6))
  sns.kdeplot(
      x=lista['lon'], y=lista['lat'],
      weights=tienda['Precio'],      # pondera por ingresos
      fill=True, cmap='magma',
      thresh=0.05, levels=100
  )
  plt.title('Mapa de Calor de Concentración del precio')
  plt.xlabel('Longitud')
  plt.ylabel('Latitud')
  plt.tight_layout()
  plt.show()

mapa_concentracion_tienda = heatmaps(tienda)
mapa_concentracion_tienda2 = heatmaps(tienda2)
mapa_concentracion_tienda3 = heatmaps(tienda3)
mapa_concentracion_tienda4 = heatmaps(tienda4)
```
